## 4 如果畫個參數式圖形

其實參數式圖形基本上也是一堆 x, y 的座標, 只是 x(t), y(t) 都是 t 的函數, 我們可以想成 t 是一條直線，我們透過點來彎折。

假設我們的

$$ -2\pi \leq t \leq 2\pi$$

要畫個半徑是 r 的圓, x(t), y(t) 就是這樣:

* $x(t) = r\cos(t)$
* $y(t) = r\sin(t)$

我們來試試事情是不是如我們想的這樣...

```python
r = 3
# np.pi= π
t = np.linspace(-2*np.pi, 2*np.pi, 200)
t
```

```python
x = r*np.cos(t)
y = r*np.sin(t)
```

```python
plt.plot(x, y, lw=3)
```

成功了... 那 `matplotlib` 怎麼畫成圓形呢? 我們來調整一下。
<br> 我們可使用 `set_aspect`
<br> `set_aspect` 是一個函數，用於設定圖形的縱橫比（aspect ratio）。縱橫比定義了圖形在 x 軸和 y 軸上的比例關係，它可以用來控制圖形的形狀和變形程度。

```python
# plt.gca是畫橢圓形，但給定參數就可以變成圓形
ax = plt.gca()
ax.set_aspect('equal') # 設定縱橫比為equal，即保持 x 軸和 y 軸的比例相等
plt.plot(x, y, lw=3)
```

再來我們可以想一下, 如果 r 不是故定的呢? 也就是它會變長變短, 是不是會畫出很多特別的圖呢? 其實很多特別的曲線不過就是找個試合的 r(t), 我們來隨意試試看。

```python
r = np.sin(5*t)
x = r*np.cos(t)
y = r*np.sin(t)

ax = plt.gca()
ax.set_aspect('equal')
plt.plot(x, y, lw=3)
```

是不是很有意思呢? 你也可以自己試試不同的變化。

## 5 `subplot` 畫多個圖

我們每次畫圖的時候, `matplotlib` 就弄 1 個 figure 畫圖區出來, 裡面可以有很多子圖, 在 figure 裡叫 axes。目前我們都只有 1 個 figure 內含 1 張圖, 所以都不用設, 現在我想畫 4 張圖時。我們就要先想好「陣式」。

比如說 2x2 這樣排列的 4 張圖。

`plt.subplot` 函數的常見用法是 plt.subplot(nrows, ncols, index)，
<br> 其中 nrows 表示子圖的行數，ncols 表示子圖的列數，index 表示當前子圖的索引位置。

以下是使用 plt.subplot 的示例：

```python
import matplotlib.pyplot as plt

# 創建一個 2x2 的子圖佈局，並選擇第一個子圖作為當前子圖
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.title('Subplot 1')

# 選擇第二個子圖作為當前子圖
plt.subplot(2, 2, 2)
plt.plot([1, 2, 3, 4], [1, 8, 27, 64])
plt.title('Subplot 2')

# 選擇第三個子圖作為當前子圖
plt.subplot(2, 2, 3)
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.title('Subplot 3')

# 選擇第四個子圖作為當前子圖
plt.subplot(2, 2, 4)
plt.plot([1, 2, 3, 4], [4, 3, 2, 1])
plt.title('Subplot 4')

plt.tight_layout()  # 調整子圖間的間距
plt.show()
```

```python
x = np.linspace(0, 10, 100)

plt.subplot(221) # 分成2X2 佔用第一個，也就是佈局中的左上角子圖
plt.plot(x, np.sin(x), c='#e63946', lw=3)

plt.subplot(222) # 分成2X2 佔用第二個，也就是佈局中的右上角子圖
plt.plot(x,np.sin(3*x), c='#7fb069', lw=3)

plt.subplot(223) # 分成2X2 佔用第三個，也就是佈局中的左下角子圖
plt.scatter(x, np.random.randn(100), c='#daa73e', s=50, alpha=0.5)

plt.subplot(224) # 分成2X2 佔用第四個，也就是佈局中的右下角子圖
plt.bar(range(10), np.random.randint(1,30,10), fc='#e55934')
```

<br> bottom：這是指定長條圖底部位置的參數。bottom 表示長條圖的起始位置，這裡使用了堆疊的方式，第一個長條圖 B 的底部從 A 開始，第二個長條圖 C 的底部從 A+B 開始

```python
A = np.random.randint(2,15,5)
B = np.random.randint(2,15,5)
C = np.random.randint(2,15,5)

plt.bar(x, A, fc='#e63946', ec='none')
plt.bar(x, B, fc='#7fb069', ec='none',bottom=A)
plt.bar(x, C, fc='#e55934', ec='none',bottom=A+B)
```

## 12 散點圖 

plt.scatter：這是 Matplotlib 中用於繪製散點圖的函數。

X、Y：分別表示散點的 x 和 y 坐標。

s=75：點的大小。s 表示點的大小（size），設置為 75。

c=T：點的顏色。

alpha=.5：透明度的參數。設置為 0.5，使點有一定的透明效果。

```python
import numpy as np
import matplotlib.pyplot as plt

n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)

plt.axes([0.025,0.025,0.95,0.95])
plt.scatter(X,Y, s=75, c=T, alpha=.5)

plt.xlim(-1.5,1.5), plt.xticks([])
plt.ylim(-1.5,1.5), plt.yticks([])

plt.show()
```

add_subplot 函數中的參數 111 表示在圖形中創建一個 1x1 的子圖，將其索引設為 1。projection='3d' 表示這是一個 3D 子圖
<br> X = np.arange(-4, 4, 0.25) 和 Y = np.arange(-4, 4, 0.25)：用於生成網格點。

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
import pandas as pd

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d') #生成3D圖 
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y) 
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)
ax.set_zlim(-2,2)
show()
```

```python
df_test=pd.DataFrame(data=np.random.normal(0,1,(20,3)),columns=['a','b','c'])
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(df_test['a'],df_test['b'],df_test['c'])
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('c')
plt.show()
```
